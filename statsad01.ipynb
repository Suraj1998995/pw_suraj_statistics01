Q1. What is a random variable in probability theory?
A random variable is a variable that takes on numerical values based on the outcome of a random event.
It helps us quantify uncertainty in probability.

Example: Tossing a coin → Let X = 1 if heads, X = 0 if tails → X is a random variable.


Q2. What are the types of random variables?
There are two main types:

Discrete Random Variable:
Takes specific, countable values.
Example: Number of heads in 3 coin tosses → {0, 1, 2, 3}

Continuous Random Variable:
Takes any value within a range (possibly infinite).
Example: The time it takes to run a mile → Can be 5.5, 5.53, etc.


Q3. What is the difference between discrete and continuous distributions?
Ans:
The difference between discrete and continuous distributions lies in the type of values the random variable can take. A discrete distribution involves variables that take on countable and distinct values, such as the number of students in a class or the outcome of rolling a die. You can list all possible outcomes. On the other hand, a continuous distribution involves variables that can take an infinite number of values within a given range, like the height or weight of students. For discrete distributions, it is possible to assign a non-zero probability to exact values (e.g., P(X = 2) = 0.3), while for continuous distributions, the probability of any single exact value is always zero—only the probability over a range makes sense. Discrete data is typically visualized using bar charts, while continuous data is shown with smooth curves like bell curves.


Q4. What are probability distribution functions (PDF)?
Ans: A Probability Distribution Function (PDF) describes how probabilities are distributed over values of a random variable.

For discrete variables, it's called a Probability Mass Function (PMF):
Example: P(X = 2) = 0.25

For continuous variables, PDF is a curve, and probabilities are areas under the curve.
Example: P(1 < X < 2) is the area between X = 1 and X = 2.


Q5. How do cumulative distribution functions (CDF) differ from PDFs?
Ans: CDF (Cumulative Distribution Function) gives the probability that a random variable is less than or equal to a certain value.

PDF shows the probability at a specific point or density.



6. What is a discrete uniform distribution?
A discrete uniform distribution is a probability distribution where all outcomes are equally likely. For example, when rolling a fair six-sided die, each outcome (1 to 6) has a probability of 1/6. The probability mass function is constant for all values.


7. What are the key properties of a Bernoulli distribution?
The Bernoulli distribution models a random experiment with only two outcomes: success (1) and failure (0). It has a single parameter p, which is the probability of success.

Mean = p

Variance = p(1 - p)
Example: Flipping a coin (Heads = 1, Tails = 0)


8. What is the binomial distribution, and how is it used in probability?
The binomial distribution models the number of successes in n independent Bernoulli trials with the same probability of success p.
It is used when you're counting how many times an event occurs in a fixed number of trials.
Example: Probability of getting 3 heads in 5 coin tosses.


9. What is the Poisson distribution and where is it applied?
The Poisson distribution models the number of times an event occurs in a fixed interval of time or space when the events occur independently and at a constant average rate.
It is used in situations like:

Number of customer arrivals per hour

Number of emails received per day
Mean and variance of a Poisson distribution are both equal to λ (lambda).



10. What is a continuous uniform distribution?
A continuous uniform distribution models a situation where all values in a continuous interval [a, b] are equally likely.


11. What are the characteristics of a normal distribution?
A normal distribution is a bell-shaped, symmetric curve centered around the mean (μ), and defined by two parameters:

Mean (μ): Center of the distribution

Standard deviation (σ): Spread of the distribution
It is used widely in real-world data (e.g., heights, test scores).


12. What is the standard normal distribution, and why is it important?
The standard normal distribution is a special case of the normal distribution with:

Mean = 0

Standard deviation = 1
It is used to calculate Z-scores and helps in comparing different normal distributions by converting them to a common scale.


13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean will be approximately normal if the sample size is large enough, regardless of the population's distribution.
It’s critical because it allows statisticians to use the normal distribution for inference even when data is not normally distributed.


14. How does the Central Limit Theorem relate to the normal distribution?
The CLT explains why the normal distribution is so commonly used in statistics. It shows that the average of a large number of independent, identically distributed variables will follow a normal distribution, enabling reliable use of tools like confidence intervals and hypothesis tests.


15. What is the application of Z statistics in hypothesis testing?
Z-statistics are used to test hypotheses about population means when the population standard deviation is known and the sample size is large.
It helps in calculating the Z-score, which measures how far a sample statistic is from the population mean in terms of standard deviations. This score is then used to determine the P-value and accept or reject the null hypothesis.

17. What are point estimates and interval estimates in statistics?
A point estimate is a single value that serves as an estimate of a population parameter (e.g., the sample mean estimates the population mean). An interval estimate, on the other hand, provides a range (typically a confidence interval) within which the population parameter is expected to lie, along with a confidence level (e.g., 95%). While point estimates are precise, interval estimates give more information about the estimate’s reliability.



18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals (CIs) are essential because they provide a range of plausible values for a population parameter based on sample data. They help quantify the uncertainty around a point estimate. A 95% confidence interval, for example, means that if we repeat the experiment many times, about 95% of the intervals calculated will contain the true population parameter. This adds reliability to data-driven decisions.


19. What is the relationship between a Z-score and a confidence interval?
Z-scores are used in calculating confidence intervals when the population standard deviation is known. For a given confidence level (e.g., 95%), a corresponding Z-score (e.g., ±1.96) is used to define the margin of error around the sample mean.


20. How are Z-scores used to compare different distributions?
Z-scores allow for comparison between values from different distributions by converting raw values into standardized scores. Since Z-scores are based on the number of standard deviations from the mean, they eliminate units and make different datasets comparable. For example, comparing exam scores from two tests with different scales becomes easy when both scores are converted to Z-scores.


21. What are the assumptions for applying the Central Limit Theorem?
The Central Limit Theorem (CLT) relies on several assumptions:

Independence: The samples must be independent of each other.

Identically distributed: The population the samples are drawn from should be identically distributed.



Finite variance: The population should have a finite standard deviation.
When these conditions are met, the sampling distribution of the mean will approximate a normal distribution.



22. What is the concept of expected value in a probability distribution?
The expected value (E[X]) is the long-run average outcome of a random variable over many trials. It represents the theoretical mean of the distribution.


23. How does a probability distribution relate to the expected outcome of a random variable?
A probability distribution describes how the probabilities are distributed over the values of a random variable. It provides the basis for computing the expected outcome (expected value), which is essentially the weighted average of all possible values, weighted by their probabilities. This makes the probability distribution a critical tool in predicting average outcomes in the long run.
